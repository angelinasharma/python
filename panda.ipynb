### What is Pandas
Pandas is a library used for working with data sets.
It has functions for analyzing, cleaning, exploring, and maniplulating data.

```python
import pandas as pd

dataset = {
    'Animals': ["Cats", "Dogs", "Bunnies"],
    'Passings': [1,2,3],
}

var = pd.DataFrame(dataset)

print(var)
```

```python
a = [2,4,7]

var2 = pd.Series(a)
var3 = pd.DataFrame(a)
print(var2)
print()
print(var3)
```

```python
a = [2,4,7]

var4 = pd.Series(a, index=["x", "y", "z"])

print(var4)
```

```python
calories = {"day 1": 420, "day 2": 380, "day 3": 390}

var6 = pd.Series(calories)
print(var6)
```

```python
calories = {"day 1": 420, "day 2": 380, "day 3": 390}

var7 = pd.Series(calories, index=["day 1", "day 2"])
print(var7)
```

```python
data = {
    "calories": [420, 380, 390],
    "duration": [50,45,40]
}

var8 = pd.DataFrame(data)
print(var8)
print()
print(var8.loc[1])
```

```python
data = {
    "calories": [420, 380, 390],
    "duration": [50,45,40]
}

df = pd.DataFrame(data, index= ["day 1", "day 2", "day 3"])

print(df)
```

```python
df2 = pd.read_csv('data.csv')

print(df2)
```
