```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([89, 78, 97, 76,83])
ypoints = np.array([240, 258, 245, 276, 298])

plt.plot(xpoints, ypoints)
plt.xlabel("Avg Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```
